```python
from framework.plots import Plotter
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
radio_image_name = "/home/miguel/Documents/galaxyclusters_FOF/A1314/A1314_I.fits"
xray_image_name = "/home/miguel/Documents/galaxyclusters_FOF/A1314/xray.fits"
```

```python
hdul_radio = fits.open(radio_image_name)
hdul_xray = fits.open(xray_image_name)
```

```python
plotter = Plotter(radio_image=hdul_radio, xray_image=hdul_xray, use_latex=True)
```

```python
plotter.get_optical(rgb=False)
```

```python
plt.imshow(plotter.optical_image.data, origin='lower')
```

```python
plotter.get_additional(surveys="WISE 3.4")
```

```python
plt.imshow(plotter.additional_image.data, origin='lower')
```

```python
plt.imshow(plotter.radio_image[0].data.squeeze(), origin='lower')
```

```python
plotter.plot_overlay(xlabel="Right Ascension [J2000]", ylabel='Declination [J2000]', sigma_list=[2.5, 50.0, 5.0])
```
